# Beauty Dataset Preprocessing

# Combining the beauty datasets together in one big dataset

```python
# Tag each brand then combine
beauty1_df["brand"] = "Nykaa"
beauty2_df["brand"] = "Purplle"
beauty3_df["brand"] = "Tira"

beauty_df = pd.concat([beauty1_df, beauty2_df, beauty3_df], ignore_index=True)
beauty_df["data_source"] = "beauty"

# Standardize column names across all datasets (lowercase, underscores)
def standardize_cols(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
    return df

beauty_df         = standardize_cols(beauty_df)
social_media_df   = standardize_cols(social_media_df)
digital_marketing_df = standardize_cols(digital_marketing_df)

print(f"Combined beauty: {beauty_df.shape}")
print(f"\nBeauty columns:  {list(beauty_df.columns)}")
print(f"\nSocial columns:  {list(social_media_df.columns)}")
print(f"\nDigital columns: {list(digital_marketing_df.columns)}")
```

# Unifying "platform" column on the beauty dataset

```python
# ── Beauty: explode multi-platform rows ──────────────────────────────
beauty_df["platform"] = beauty_df["channel_used"].astype(str).str.split(",")
beauty_df = beauty_df.explode("platform")
beauty_df["platform"] = beauty_df["platform"].str.strip().str.lower()

BEAUTY_MAP = {
    "instagram": "Instagram", "youtube": "YouTube",
    "whatsapp": "WhatsApp",   "google": "Google",
    "email": "Email",         "facebook": "Facebook",
    "twitter": "Twitter",     "linkedin": "LinkedIn",
    "snapchat": "Snapchat",   "pinterest": "Pinterest",
}
beauty_df["platform"] = beauty_df["platform"].map(BEAUTY_MAP).fillna("Other")
beauty_df.drop(columns=["channel_used"], inplace=True)

# ── Social media: already clean platform names ────────────────────────
social_media_df["platform"] = social_media_df["channel_used"].str.strip().str.title()
social_media_df["data_source"] = "social_media"
social_media_df.drop(columns=["channel_used"], inplace=True)

# ── Digital: map channel types → platform names ───────────────────────
DIGITAL_MAP = {
    "Search": "Google", "Video": "YouTube",
    "Email": "Email",   "Display": "Display",
    "Social": "Other_Social",
}
digital_marketing_df["platform"] = digital_marketing_df["channel"].map(DIGITAL_MAP).fillna("Other")
digital_marketing_df["data_source"] = "digital"
digital_marketing_df.drop(columns=["channel"], inplace=True)

print(f"Beauty after explode:  {len(beauty_df):,} rows")
print(f"\nBeauty platforms:\n{beauty_df['platform'].value_counts()}")
print(f"\nSocial platforms:\n{social_media_df['platform'].value_counts()}")
print(f"\nDigital platforms:\n{digital_marketing_df['platform'].value_counts()}")
```

# Standardize and align coumns across all the datasets before merging

```python
# ── Beauty: rename to unified schema ─────────────────────────────────
beauty_df = beauty_df.rename(columns={
    "duration": "duration_days",
})

# ── Social: rename to unified schema ─────────────────────────────────
social_media_df = social_media_df.rename(columns={
    "campaign_goal": "campaign_type",
    "company":       "brand",
    "duration":      "duration_days",
    "location":      "geo_location",
})

# ── Digital: rename to unified schema ────────────────────────────────
digital_marketing_df = digital_marketing_df.rename(columns={
    "spend_usd":   "acquisition_cost",
    "revenue_usd": "revenue",
    "conversion":  "conversions",
    "segment":     "customer_segment",
    "country":     "geo_location",
})

# ── Fix types: social acquisition_cost may be a currency string ───────
if social_media_df["acquisition_cost"].dtype == object:
    social_media_df["acquisition_cost"] = (
        social_media_df["acquisition_cost"]
        .str.replace(r"[^\d.]", "", regex=True)
        .astype(float)
    )

# ── Fix types: social duration may be "X days" string ─────────────────
if social_media_df["duration_days"].dtype == object:
    social_media_df["duration_days"] = (
        social_media_df["duration_days"]
        .str.extract(r"(\d+)")[0]
        .astype(float)
    )

print("Beauty dtypes:")
print(beauty_df[["impressions","clicks","conversions","acquisition_cost","roi","duration_days"]].dtypes)
print("\nSocial dtypes:")
print(social_media_df[["impressions","clicks","conversion_rate","acquisition_cost","roi","duration_days"]].dtypes)
print("\nDigital dtypes:")
print(digital_marketing_df[["impressions","clicks","conversions","acquisition_cost","roi"]].dtypes)
```

# Feature Engineering the General features

```python
def safe_divide(num, den, fill=0.0):
    return num.div(den.replace(0, np.nan)).fillna(fill)

# ── Beauty ────────────────────────────────────────────────────────────
beauty_df["ctr"]             = safe_divide(beauty_df["clicks"], beauty_df["impressions"])
beauty_df["conversion_rate"] = safe_divide(beauty_df["conversions"], beauty_df["clicks"])
beauty_df["cpc"]             = safe_divide(beauty_df["acquisition_cost"], beauty_df["clicks"])
beauty_df["cpm"]             = safe_divide(beauty_df["acquisition_cost"], beauty_df["impressions"]) * 1000
beauty_df["cpr"]             = safe_divide(beauty_df["acquisition_cost"], beauty_df["conversions"].clip(lower=1))
multiplier                   = np.random.uniform(1.2, 2.5, size=len(beauty_df))
beauty_df["budget"]          = (beauty_df["acquisition_cost"] * beauty_df["conversions"].clip(lower=1) * multiplier).round(2)

# ── Social ────────────────────────────────────────────────────────────
social_media_df["ctr"]         = safe_divide(social_media_df["clicks"], social_media_df["impressions"])
social_media_df["cpc"]         = safe_divide(social_media_df["acquisition_cost"], social_media_df["clicks"])
social_media_df["cpm"]         = safe_divide(social_media_df["acquisition_cost"], social_media_df["impressions"]) * 1000
# conversions estimated from conversion_rate × clicks
social_media_df["conversions"] = (social_media_df["conversion_rate"] * social_media_df["clicks"]).round(0).astype(int)
social_media_df["cpr"]         = safe_divide(social_media_df["acquisition_cost"], social_media_df["conversions"].clip(lower=1))
multiplier                     = np.random.uniform(1.2, 2.5, size=len(social_media_df))
social_media_df["budget"]      = (social_media_df["acquisition_cost"] * multiplier).round(2)

# ── Digital ───────────────────────────────────────────────────────────
digital_marketing_df["ctr"]             = safe_divide(digital_marketing_df["clicks"], digital_marketing_df["impressions"])
digital_marketing_df["conversion_rate"] = safe_divide(digital_marketing_df["conversions"], digital_marketing_df["clicks"])
digital_marketing_df["cpc"]             = safe_divide(digital_marketing_df["acquisition_cost"], digital_marketing_df["clicks"])
digital_marketing_df["cpm"]             = safe_divide(digital_marketing_df["acquisition_cost"], digital_marketing_df["impressions"]) * 1000
digital_marketing_df["cpr"]             = safe_divide(digital_marketing_df["acquisition_cost"], digital_marketing_df["conversions"].clip(lower=1))
multiplier                              = np.random.uniform(1.2, 2.5, size=len(digital_marketing_df))
digital_marketing_df["budget"]          = (digital_marketing_df["acquisition_cost"] * multiplier).round(2)

print("✓ Beauty derived features:")
print(beauty_df[["ctr","conversion_rate","cpc","cpm","cpr","budget"]].describe().round(4))
print("\n✓ Social derived features:")
print(social_media_df[["ctr","conversion_rate","cpc","cpm","cpr","budget"]].describe().round(4))
print("\n✓ Digital derived features:")
print(digital_marketing_df[["ctr","conversion_rate","cpc","cpm","cpr","budget"]].describe().round(4))
```

# Generate Platorm-Specific Features

```python
rng = np.random.default_rng(42)

def generate_platform_specific(df):
    n = len(df)
    platform = df["platform"]

    # ── Video views (TikTok, YouTube, Instagram only) ─────────────────
    # video_view_rate: % of impressions that became video views
    # TikTok: 0.15–0.60, YouTube: 0.20–0.70, Instagram: 0.10–0.45
    video_view_rate = np.where(
        platform == "TikTok",   rng.uniform(0.15, 0.60, n),
        np.where(
        platform == "YouTube",  rng.uniform(0.20, 0.70, n),
        np.where(
        platform == "Instagram",rng.uniform(0.10, 0.45, n),
        np.nan)))

    video_views = np.where(
        ~np.isnan(video_view_rate),
        (df["impressions"] * video_view_rate).round(0),
        np.nan)

    # ── Frequency (Facebook, TikTok, Instagram, Pinterest only) ───────
    # Frequency = impressions / reach, typically 1.2–4.5
    frequency = np.where(
        platform.isin(["Facebook", "TikTok", "Instagram", "Pinterest"]),
        rng.uniform(1.2, 4.5, n).round(2),
        np.nan)

    # ── Reach (derived from impressions / frequency) ───────────────────
    reach = np.where(
        ~np.isnan(frequency),
        (df["impressions"] / frequency).round(0),
        np.nan)

    # ── Purchases (Facebook, Instagram, TikTok, Google only) ──────────
    # Purchases ≈ subset of conversions (0.3–0.8 of conversions)
    purchases = np.where(
        platform.isin(["Facebook", "Instagram", "TikTok", "Google"]),
        (df["conversions"] * rng.uniform(0.3, 0.8, n)).round(0),
        np.nan)

    df = df.copy()
    df["video_views"]     = video_views
    df["video_view_rate"] = video_view_rate
    df["frequency"]       = frequency
    df["reach"]           = reach
    df["purchases"]       = purchases

    return df

beauty_df          = generate_platform_specific(beauty_df)
social_media_df    = generate_platform_specific(social_media_df)
digital_marketing_df = generate_platform_specific(digital_marketing_df)

# Quick check — non-null counts per feature per dataset
for name, df in [("Beauty", beauty_df), ("Social", social_media_df), ("Digital", digital_marketing_df)]:
    print(f"\n{name} — platform-specific feature coverage:")
    cols = ["platform","video_views","video_view_rate","frequency","reach","purchases"]
    print(df[cols].groupby("platform").count())
```

# Generating the "TikTok" rows since they don't exist

```python
def generate_tiktok(n=15000):
    rng2 = np.random.default_rng(99)

    impressions      = rng2.integers(5_000, 500_000, size=n)
    ctr              = rng2.uniform(0.01, 0.04, n)
    clicks           = (impressions * ctr).astype(int).clip(min=1)
    conversion_rate  = rng2.uniform(0.005, 0.06, n)
    conversions      = (clicks * conversion_rate).astype(int).clip(min=0)
    cpc              = rng2.uniform(0.10, 1.20, n)
    acquisition_cost = (clicks * cpc).round(2)
    cpm              = (acquisition_cost / impressions * 1000).round(4)
    cpr              = np.where(conversions > 0, acquisition_cost / conversions, 0.0).round(4)
    revenue          = (conversions * rng2.uniform(5, 150, n)).round(2)
    roi              = np.where(
                           acquisition_cost > 0,
                           (revenue - acquisition_cost) / acquisition_cost,
                           0.0).round(4)
    budget           = (acquisition_cost * rng2.uniform(1.2, 2.5, n)).round(2)
    engagement_score = rng2.uniform(0.3, 0.95, n).round(4)

    # Platform-specific
    video_view_rate  = rng2.uniform(0.15, 0.60, n).round(4)
    video_views      = (impressions * video_view_rate).round(0)
    frequency        = rng2.uniform(1.2, 4.5, n).round(2)
    reach            = (impressions / frequency).round(0)
    purchases        = (conversions * rng2.uniform(0.3, 0.8, n)).round(0)
    leads            = (conversions * rng2.uniform(0.1, 0.4, n)).round(0)

    campaign_types   = ["Video", "Brand Awareness", "Engagement", "Conversion", "Product Launch"]
    audiences        = ["Gen Z", "Millennials", "18-24", "25-34", "Young Adults"]
    segments         = ["Gen Z", "Millennials", "Young Adults", "Entertainment", "Gaming"]
    languages        = ["English", "Arabic", "Spanish", "French", "Hindi"]
    statuses         = ["Active", "Paused", "Completed"]

    dates = pd.date_range("2021-01-01", "2023-12-31", periods=n)

    tiktok_df = pd.DataFrame({
        "campaign_id"      : [f"TT_{i:06d}" for i in range(n)],
        "campaign_type"    : rng2.choice(campaign_types, size=n),
        "target_audience"  : rng2.choice(audiences, size=n),
        "duration_days"    : rng2.integers(3, 60, size=n),
        "platform"         : "TikTok",
        "impressions"      : impressions,
        "clicks"           : clicks,
        "conversions"      : conversions,
        "revenue"          : revenue,
        "acquisition_cost" : acquisition_cost,
        "roi"              : roi,
        "language"         : rng2.choice(languages, size=n),
        "engagement_score" : engagement_score,
        "customer_segment" : rng2.choice(segments, size=n),
        "date"             : dates,
        "ctr"              : ctr.round(6),
        "conversion_rate"  : conversion_rate.round(6),
        "cpc"              : cpc.round(4),
        "cpm"              : cpm,
        "cpr"              : cpr,
        "budget"           : budget,
        "leads"            : leads,
        "purchases"        : purchases,
        "video_views"      : video_views,
        "video_view_rate"  : video_view_rate,
        "frequency"        : frequency,
        "reach"            : reach,
        "brand"            : "Unknown",
        "data_source"      : "synthetic_tiktok",
    })
    return tiktok_df

tiktok_df = generate_tiktok(n=15000)

print(f"TikTok rows: {len(tiktok_df):,}")
print(f"\nSample stats:")
print(tiktok_df[["impressions","clicks","ctr","conversion_rate","roi","video_view_rate","frequency"]].describe().round(4))
```

# Unifying everything in one dataset

```python
FINAL_COLUMNS = [
    # Identity
    "campaign_id", "data_source", "brand", "platform", "date",
    # Setup
    "campaign_type", "target_audience", "customer_segment",
    "language", "duration_days",
    # Spend
    "acquisition_cost", "budget",
    # Reach & delivery
    "impressions", "reach", "frequency",
    # Engagement
    "clicks", "ctr", "cpc", "cpm", "engagement_score",
    # Video
    "video_views", "video_view_rate",
    # Conversion
    "conversions", "conversion_rate", "cpr",
    "leads", "purchases",
    # Revenue
    "revenue", "roi",
]

# Add missing columns as NaN before concat
def align_schema(df, columns):
    for col in columns:
        if col not in df.columns:
            df[col] = np.nan
    return df[columns].copy()

beauty_aligned  = align_schema(beauty_df, FINAL_COLUMNS)
social_aligned  = align_schema(social_media_df, FINAL_COLUMNS)
digital_aligned = align_schema(digital_marketing_df, FINAL_COLUMNS)
tiktok_aligned  = align_schema(tiktok_df, FINAL_COLUMNS)

unified_df = pd.concat(
    [beauty_aligned, social_aligned, digital_aligned, tiktok_aligned],
    ignore_index=True
)

# Parse dates
unified_df["date"] = pd.to_datetime(unified_df["date"], errors="coerce")

# Add time features
unified_df["year"]        = unified_df["date"].dt.year
unified_df["month"]       = unified_df["date"].dt.month
unified_df["quarter"]     = unified_df["date"].dt.quarter
unified_df["day_of_week"] = unified_df["date"].dt.dayofweek

# Clip top 0.1% outliers on numeric spend/volume columns
for col in ["impressions", "clicks", "conversions", "revenue", "budget", "acquisition_cost"]:
    cap = unified_df[col].quantile(0.999)
    unified_df[col] = unified_df[col].clip(upper=cap)

print(f"Unified dataset: {unified_df.shape}")
print(f"\nPlatform distribution:\n{unified_df['platform'].value_counts()}")
print(f"\nData source distribution:\n{unified_df['data_source'].value_counts()}")
print(f"\nMissing values:\n{unified_df.isnull().sum()[unified_df.isnull().sum() > 0]}")
```

# Fixing post-merge issues

```python
# ── Fix 1: social date was object type, never parsed before align ──────
# Re-add date back to social_aligned from social_media_df directly
social_aligned["date"] = pd.to_datetime(social_media_df["date"], errors="coerce").values

# ── Fix 2: impute digital missing columns with reasonable values ────────
# These are campaign-level fields the digital dataset simply didn't track

# brand → "Unknown" (same as TikTok synthetic)
digital_aligned["brand"] = "Unknown"

# campaign_type → impute from platform (Search=Performance, Social=Engagement, etc.)
campaign_type_map = {
    "Google"      : "Performance",
    "YouTube"     : "Video",
    "Email"       : "Email Marketing",
    "Display"     : "Display",
    "Other_Social": "Engagement",
}
digital_aligned["campaign_type"] = digital_marketing_df["platform"].map(campaign_type_map).values

# target_audience → "General" (digital dataset had no audience info)
digital_aligned["target_audience"] = "General"

# language → "English" (dataset is US/global, English is safe default)
digital_aligned["language"] = "English"

# duration_days → median of other datasets (~30 days is typical)
digital_aligned["duration_days"] = 30

# engagement_score → derive from ctr (normalized to 0-1 range)
# higher CTR → higher engagement, scaled to realistic 0.3–0.9 range
ctr_vals = digital_aligned["ctr"].fillna(0)
eng_min, eng_max = 0.3, 0.9
ctr_min, ctr_max = ctr_vals.min(), ctr_vals.max()
digital_aligned["engagement_score"] = (
    eng_min + (ctr_vals - ctr_min) / (ctr_max - ctr_min + 1e-9) * (eng_max - eng_min)
).round(4)

# ── Re-unify ───────────────────────────────────────────────────────────
unified_df = pd.concat(
    [beauty_aligned, social_aligned, digital_aligned, tiktok_aligned],
    ignore_index=True
)

unified_df["date"]        = pd.to_datetime(unified_df["date"], errors="coerce")
unified_df["year"]        = unified_df["date"].dt.year
unified_df["month"]       = unified_df["date"].dt.month
unified_df["quarter"]     = unified_df["date"].dt.quarter
unified_df["day_of_week"] = unified_df["date"].dt.dayofweek

for col in ["impressions","clicks","conversions","revenue","budget","acquisition_cost"]:
    cap = unified_df[col].quantile(0.999)
    unified_df[col] = unified_df[col].clip(upper=cap)

print(f"Unified dataset: {unified_df.shape}")
print(f"\nMissing values after fixes:")
missing = unified_df.isnull().sum()
missing = missing[missing > 0]
print(missing)
```

```python
# ── Fix 1: digital dataset has no dates → assign random dates ─────────
# Use same range as other datasets: 2021–2023
digital_mask = unified_df["data_source"] == "digital"
n_digital = digital_mask.sum()

rng3 = np.random.default_rng(7)
random_dates = pd.to_datetime(
    rng3.integers(
        pd.Timestamp("2021-01-01").value,
        pd.Timestamp("2023-12-31").value,
        size=n_digital
    )
)
unified_df.loc[digital_mask, "date"] = random_dates
unified_df.loc[digital_mask, "year"]        = random_dates.year
unified_df.loc[digital_mask, "month"]       = random_dates.month
unified_df.loc[digital_mask, "quarter"]     = random_dates.quarter
unified_df.loc[digital_mask, "day_of_week"] = random_dates.dayofweek

# ── Fix 2: social_media_df has no revenue → estimate from ROI ─────────
# ROI = (revenue - cost) / cost  →  revenue = cost × (1 + ROI)
social_mask = unified_df["data_source"] == "social_media"
unified_df.loc[social_mask, "revenue"] = (
    unified_df.loc[social_mask, "acquisition_cost"] *
    (1 + unified_df.loc[social_mask, "roi"])
).round(2)

# ── Final missing check ────────────────────────────────────────────────
print("Missing values after all fixes:")
missing = unified_df.isnull().sum()
missing = missing[missing > 0]
print(missing)

print(f"\nFinal shape: {unified_df.shape}")
print(f"\nDate range: {unified_df['date'].min()} → {unified_df['date'].max()}")
print(f"\nRevenue sample (social):")
print(unified_df.loc[social_mask, ["acquisition_cost","roi","revenue"]].head())
```

```python
# ── Fix date cap: clip anything beyond 2023-12-31 ─────────────────────
unified_df["date"] = unified_df["date"].clip(upper=pd.Timestamp("2023-12-31"))
unified_df["year"]        = unified_df["date"].dt.year
unified_df["month"]       = unified_df["date"].dt.month
unified_df["quarter"]     = unified_df["date"].dt.quarter
unified_df["day_of_week"] = unified_df["date"].dt.dayofweek

print(f"Date range after fix: {unified_df['date'].min()} → {unified_df['date'].max()}")

# ── Performance label ──────────────────────────────────────────────────
# Composite score = 0.4×ROI_norm + 0.3×CTR_norm + 0.3×CVR_norm
# Computed globally so percentiles are cross-platform

def normalize(s):
    mn, mx = s.min(), s.max()
    return (s - mn) / (mx - mn + 1e-9)

score = (
    0.4 * normalize(unified_df["roi"]) +
    0.3 * normalize(unified_df["ctr"]) +
    0.3 * normalize(unified_df["conversion_rate"].fillna(0))
)

unified_df["performance_score"] = score.round(4)
unified_df["performance_label"] = pd.cut(
    score,
    bins=[0, 0.33, 0.66, 1.01],
    labels=["Low", "Medium", "High"],
    include_lowest=True,
)

# ── Alert flag ─────────────────────────────────────────────────────────
# Flag = 1 if underperforming on ≥2 of: ROI, CTR, CVR (below 25th percentile)
roi_low  = unified_df["roi"] < unified_df["roi"].quantile(0.25)
ctr_low  = unified_df["ctr"] < unified_df["ctr"].quantile(0.25)
cvr_low  = unified_df["conversion_rate"].fillna(0) < unified_df["conversion_rate"].fillna(0).quantile(0.25)

unified_df["alert_flag"] = (
    roi_low.astype(int) + ctr_low.astype(int) + cvr_low.astype(int) >= 2
).astype(int)

print(f"\nPerformance label distribution:")
print(unified_df["performance_label"].value_counts())
print(f"\nAlert flag distribution:")
print(unified_df["alert_flag"].value_counts())
print(f"\nPerformance score stats:")
print(unified_df["performance_score"].describe().round(4))
```

```python
# ── Recompute performance label using quantile-based bins ─────────────
low_thresh  = unified_df["performance_score"].quantile(0.33)
high_thresh = unified_df["performance_score"].quantile(0.66)

print(f"Thresholds → Low: <{low_thresh:.4f}, Medium: {low_thresh:.4f}–{high_thresh:.4f}, High: >{high_thresh:.4f}")

unified_df["performance_label"] = pd.cut(
    unified_df["performance_score"],
    bins=[
        unified_df["performance_score"].min() - 1e-9,
        low_thresh,
        high_thresh,
        unified_df["performance_score"].max() + 1e-9,
    ],
    labels=["Low", "Medium", "High"],
)

print(f"\nPerformance label distribution:")
print(unified_df["performance_label"].value_counts())

print(f"\nMean score per label:")
print(unified_df.groupby("performance_label", observed=True)["performance_score"].mean().round(4))

print(f"\nMean ROI per label:")
print(unified_df.groupby("performance_label", observed=True)["roi"].mean().round(4))
```

```python
# ── Drop columns not needed for modeling or dashboard ─────────────────
unified_df.drop(columns=["performance_score"], inplace=True)

# ── Encode performance_label as integer for modeling convenience ───────
label_map = {"Low": 0, "Medium": 1, "High": 2}
unified_df["performance_label_enc"] = unified_df["performance_label"].map(label_map)

# ── Final column order ─────────────────────────────────────────────────
FINAL_ORDER = [
    # Identity
    "campaign_id", "data_source", "brand", "platform", "date",
    "year", "month", "quarter", "day_of_week",
    # Setup
    "campaign_type", "target_audience", "customer_segment",
    "language", "duration_days",
    # Spend
    "acquisition_cost", "budget",
    # Reach & delivery
    "impressions", "reach", "frequency",
    # Engagement
    "clicks", "ctr", "cpc", "cpm", "engagement_score",
    # Video
    "video_views", "video_view_rate",
    # Conversion
    "conversions", "conversion_rate", "cpr",
    "leads", "purchases",
    # Revenue
    "revenue", "roi",
    # Target
    "performance_label", "performance_label_enc", "alert_flag",
]
unified_df = unified_df[FINAL_ORDER]

# ── Final sanity check ─────────────────────────────────────────────────
print(f"Final shape: {unified_df.shape}")
print(f"\nColumn list:\n{unified_df.columns.tolist()}")
print(f"\nDtypes:\n{unified_df.dtypes}")
print(f"\nPlatform distribution:\n{unified_df['platform'].value_counts()}")
print(f"\nRemaining nulls:\n{unified_df.isnull().sum()[unified_df.isnull().sum() > 0]}")
```

```python
# ── Fix dtypes ─────────────────────────────────────────────────────────
unified_df["performance_label_enc"] = unified_df["performance_label_enc"].astype(int)
unified_df["duration_days"]         = unified_df["duration_days"].astype(int)

# ── Verify ─────────────────────────────────────────────────────────────
print(unified_df[["duration_days","performance_label_enc"]].dtypes)
print(f"\nSample:")
print(unified_df[["platform","roi","ctr","conversion_rate","performance_label","performance_label_enc","alert_flag"]].sample(8, random_state=1))
```
